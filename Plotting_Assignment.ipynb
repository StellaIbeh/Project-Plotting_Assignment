```python
# YOUR CODE HERE
# Plot a histogram

# Set the figure size
plt.figure(figsize=(12, 8))

# Plot the histogram with different bin widths
plt.hist(data['co2'], bins=30, color='black', alpha=0.5,edgecolor="white")
plt.title('co2 Emission distribution')
plt.xlabel('Carbon Dioxide  (g/mi)')
plt.ylabel('Frequency of co2 emission')
plt.savefig('co2_vehicles')
plt.show()
```

The plot reveals the spread and concentration of emissions, aiding in understanding their potential environmental impact.

Create A HeatMap

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.xticks(rotation=45)
plt.title('Correlation Heatmap')
plt.savefig('heatmap')
plt.show()
```

Interpretation of HeatMap
The codes selects numeric data while including both floating and interger types.The  colored-cell colors represent the various correlation matrix visually . There exist both positive and negative correlation between the data sets . The colormap 'coolwarm' highlights positive and negative correlations effectively. Additionally, the fmt=".2f" ensures that correlation coefficients are displayed as floating-point numbers with two decimal places.
